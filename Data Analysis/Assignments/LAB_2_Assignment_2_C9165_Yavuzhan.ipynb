## 1a: Box plot of age in both genders separeted by manner of death.

```python
np.array(kill['manner_of_death'].value_counts().index)
```

**INSTRUCTION: Demonstrate the Qurtiles of age in both genders separeted by manner of death.**

```python
sns.boxplot(x="gender", y='age',data=kill,hue = "manner_of_death")
```

## 1b: Swarm Plot 

**INSTRUCTION: Demonstrate the age in both genders separeted by manner of death by a swarm plot.**

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x = "gender", y = "age",data = kill,hue = "manner_of_death",size = 5)
```

## 2: Plot 12 most dangerous cities

```python
dang = kill['city'].value_counts().head(12)
```

```python
plt.figure(figsize=(12,8))
plt.title('Most Dangerous Cities',color ='blue',size = 15)
sns.barplot(x = dang.index, y = list(dang), data = kill, ci = 50);
plt.xticks(rotation=45);
```

```python
# with seaborn barplot

```

```python
plt.figure(figsize=(12,6))
plt.title('Most Dangerous Cities',color ='blue',size = 15)
sns.countplot(x = 'city', data =kill, order = kill['city'].value_counts().head(12).index);
plt.xticks(rotation=45);
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states
